# Задание 2

Загрузите данные train.csv, оставьте в данных только признаки 'fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due', затем избавьтесь от пропусков. Целевой переменной будет 'balance_due', разделите данные на обучающую и тестовую выборки в соотношении 70% / 30% без перемешивания. Обучите линейную регрессию из scikit-learn и запишите в переменную result значение метрики RMSE на тестовой выборке. 


```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('./train.csv')
data.head()
```

```python
data = data.loc[:, ['fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due']]
data = data.dropna()
data.head()
```

```python
y = data['balance_due']
X = data.drop(['balance_due'], 1)
x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, shuffle=False)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_valid)
```

```python
print('RMSE =', np.sqrt(mean_squared_error(y_valid,y_pred)))
```

Целевой переменной будет 'compliance', разделите данные на обучающую и тестовую выборки в соотношении 70%/30% без перемешивания. Вычтите из значения метрики на тренировочной выборки значение метрики на тестовой выборке и запишите результат в переменную result. 

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
```

```python
data = pd.read_csv('./train.csv')
data.head()
```

```python
data = data.loc[:, ['fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due', 'compliance']]
data = data.dropna()
data.head()
```

```python
y = data['compliance']
X = data.drop(['compliance'], 1)
x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, shuffle=False)
```

```python
tree = DecisionTreeClassifier(random_state=23)
```

```python
tree.fit(x_train,y_train)
train_pred = tree.predict(x_train)
train_f1_score = f1_score(y_train,train_pred)
print('f1_score для train =',train_f1_score)

tree.fit(x_valid,y_valid)
test_pred = tree.predict(x_valid)
test_f1_score = f1_score(y_valid,test_pred)
print('f1_score для train =',test_f1_score)
```

```python
result = train_f1_score - test_f1_score
result
```

```python
data = pd.read_csv('./train.csv')
data.head()
```

```python
data = data.loc[:, ['fine_amount', 'state_fee', 'late_fee', 'discount_amount', 'balance_due', 'compliance']]
data = data.dropna()
data.head()
```

```python
y = data['compliance']
X = data.drop(['compliance'], 1)
x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, shuffle=False)
```

```python
unique_compliance, counts_compliance = np.unique(data['compliance'], return_counts=True)
print(dict(zip(unique_compliance, counts_compliance)))

min_count = min(counts_compliance)
```

```python
balanced_data = pd.DataFrame([])

for i in range(unique_compliance.shape[0]):
    new_data = data.loc[data['compliance'] == i][:min_count]
    balanced_data = pd.concat([balanced_data, new_data])

balanced_data
```

```python
y = balanced_data['compliance']
X = balanced_data.drop(['compliance'], 1)
x_train, x_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, shuffle=False)
```

```python
tree = DecisionTreeClassifier(random_state=23)
```

```python
tree.fit(x_train,y_train)
train_pred = tree.predict(x_train)
train_f1_score = f1_score(y_train,train_pred)
print('f1_score для train =',train_f1_score)
```
